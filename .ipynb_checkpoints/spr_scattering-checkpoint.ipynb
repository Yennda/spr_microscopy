```python
%matplotlib notebook
import numpy as np
from numpy.linalg import norm as norm
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm
import math as m

import matplotlib.pyplot as plt
import np_info_analysis as na
```

```python
# def U(k, x):
#     return U0*m.cos(np.dot(k, x))

# def Udiff(k, r, x):
#     return U(k, r)*m.cos(norm(x-r)*norm(k))
def U(k, x):
    return U0*np.exp(1j*np.dot(np.real(k), x))

def Udiff(k, r, x):
    rr=norm(x-r)
    
    
    p=np.dot(x-r, np.array([1,0]))/rr
    return 0.2*U(k, r)*np.exp(1j*rr*norm(np.real(k)))*p*np.exp(-norm(np.imag(k))*rr) #/(rr**0.5)


def intensity(u):
    return u*np.conjugate(u)
```

```python
U0=1
# k=np.array([2*m.pi/534e-9,0])
# k=np.array([2*m.pi/534e-9,0]) #750
# k=np.array([2*m.pi/402e-9,0])  #600

k=np.array([(11755041.21345882+78217.09878973379j), 0])
# k=np.array([(15604823.125390332-526208.287757275j), 0])

print(np.exp(78))

print(norm(np.imag(k)))
      
r=np.array([10,5])*1e-6
```

```python
X=np.arange(0,20, 20/200)*1e-6
Y=np.arange(0,10, 10/100)*1e-6
data=np.zeros([100,200])
for i in range(100):
    for j in range(200):
        x=X[j]
        y=Y[i]
        data[i,j]=intensity(U(k,np.array([x,y]))+U(-k,np.array([x,y]))+Udiff(k, r, [x,y])*np.exp(1j*m.pi))
#         data[i,j]=intensity(U(k,np.array([x,y]))+Udiff(k, r, [x,y])*np.exp(1j*m.pi))
#         data[i,j]=intensity(Udiff(k, r, [x,y])*np.exp(1j*m.pi))


#         data[i,j]=intensity(U(k,np.array([x,y]))+U(-k,np.array([x,y]))*np.exp(1j*m.pi/3.5))

```

```python

fig, ax = plt.subplots()
img = ax.imshow(data)
fig.colorbar(img)

fontprops = fm.FontProperties(size=10)
scalebar = AnchoredSizeBar(ax.transData,
           50, '50 $\mu m$', 'lower right', 
           pad=0.1,
           color='white',
           frameon=False,
           size_vertical=1,
           fontproperties=fontprops)

ax.add_artist(scalebar)
# ,clim=(0, 20)
```
